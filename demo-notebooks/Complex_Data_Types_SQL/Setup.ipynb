# Demo Setup

This notebook creates the schema and tables used in **Demo: Working with Complex Data Types in SQL**.

```python
# Get current user and set up schema name
import re

username = spark.sql("SELECT current_user()").first()[0]
clean_username = re.sub(r'[^a-zA-Z0-9]', '_', username.split('@')[0])
catalog_name = spark.sql("SELECT current_catalog()").first()[0]
schema_name = f"demo_complex_types_{clean_username}"

# Create the schema
spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog_name}`.`{schema_name}` COMMENT 'Demo schema for complex data types in SQL'")
spark.sql(f"USE CATALOG `{catalog_name}`")
spark.sql(f"USE SCHEMA `{schema_name}`")
```

```python
# Create customer_profiles table with STRUCT columns
# This demonstrates properly typed nested data (dot notation)
spark.sql("""
CREATE OR REPLACE TABLE customer_profiles (
  customer_id INT,
  name STRING,
  address STRUCT<street: STRING, city: STRING, state: STRING, zipcode: STRING>,
  preferences STRUCT<favorite_category: STRING, notifications_enabled: BOOLEAN, loyalty_tier: STRING>,
  tags ARRAY<STRING>
)
""")

# Insert sample data
spark.sql("""
INSERT INTO customer_profiles VALUES
  (1, 'Sarah Johnson', struct('123 Main St', 'Portland', 'OR', '97201'), struct('Electronics', true, 'Gold'), array('vip', 'early-adopter')),
  (2, 'James Chen', struct('456 Oak Ave', 'Seattle', 'WA', '98101'), struct('Home & Garden', true, 'Silver'), array('seasonal-buyer')),
  (3, 'Maria Garcia', struct('789 Pine Rd', 'Denver', 'CO', '80201'), struct('Sports', false, 'Bronze'), array('budget-conscious', 'new-customer')),
  (4, 'Robert Kim', struct('321 Elm Blvd', 'Austin', 'TX', '73301'), struct('Electronics', true, 'Gold'), array('vip', 'tech-enthusiast')),
  (5, 'Lisa Patel', struct('654 Cedar Ln', 'Chicago', 'IL', '60601'), struct('Clothing', true, 'Silver'), array('fashion-forward')),
  (6, 'David Wilson', struct('987 Birch Dr', 'Boston', 'MA', '02101'), struct('Food & Beverage', false, 'Bronze'), array('new-customer')),
  (7, 'Emma Thompson', struct('147 Maple Way', 'San Francisco', 'CA', '94101'), struct('Electronics', true, 'Platinum'), array('vip', 'early-adopter', 'tech-enthusiast')),
  (8, 'Michael Brown', struct('258 Walnut St', 'Nashville', 'TN', '37201'), struct('Sports', true, 'Gold'), array('fitness-enthusiast'))
""")

count = spark.sql("SELECT COUNT(*) FROM customer_profiles").first()[0]
```

```python
# Create raw_events table with JSON string column
# This demonstrates raw/untyped JSON that needs parsing
spark.sql("""
CREATE OR REPLACE TABLE raw_events (
  event_id INT,
  event_type STRING,
  event_timestamp TIMESTAMP,
  payload STRING
)
""")

# Insert events with raw JSON payloads (different structures per event type)
spark.sql("""
INSERT INTO raw_events VALUES
  (1, 'purchase', '2024-03-15 10:30:00', '{"product_id": 101, "product_name": "Wireless Headphones", "amount": 79.99, "currency": "USD", "quantity": 1}'),
  (2, 'purchase', '2024-03-15 11:45:00', '{"product_id": 205, "product_name": "Running Shoes", "amount": 129.50, "currency": "USD", "quantity": 1}'),
  (3, 'page_view', '2024-03-15 09:00:00', '{"page": "/products/electronics", "referrer": "google.com", "duration_seconds": 45}'),
  (4, 'purchase', '2024-03-15 14:20:00', '{"product_id": 342, "product_name": "Garden Tools Set", "amount": 54.99, "currency": "USD", "quantity": 2}'),
  (5, 'page_view', '2024-03-15 08:15:00', '{"page": "/deals", "referrer": "email-campaign", "duration_seconds": 120}'),
  (6, 'signup', '2024-03-15 16:00:00', '{"plan": "premium", "trial_days": 14, "source": "landing-page"}'),
  (7, 'purchase', '2024-03-15 17:30:00', '{"product_id": 118, "product_name": "Bluetooth Speaker", "amount": 45.00, "currency": "EUR", "quantity": 3}'),
  (8, 'page_view', '2024-03-15 12:00:00', '{"page": "/checkout", "referrer": "internal", "duration_seconds": 30}'),
  (9, 'signup', '2024-03-15 19:00:00', '{"plan": "basic", "trial_days": 7, "source": "social-media"}'),
  (10, 'purchase', '2024-03-15 20:15:00', '{"product_id": 501, "product_name": "Yoga Mat", "amount": 35.00, "currency": "USD", "quantity": 1}')
""")

count = spark.sql("SELECT COUNT(*) FROM raw_events").first()[0]
```
