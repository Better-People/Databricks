# DEMO: Build Dashboard Visualizations in the UI

Following your preferences, this notebook is now focused on the learner workflow in the dashboard UI.

This demo assumes you already completed the dashboard datasets demo and already have a dashboard with these datasets:
* `star_schema_region_summary`
* `gold_sales`
* `vw_monthly_revenue`
* `vw_region_performance`
* `vw_top_products`
* `mv_sales_metrics`

Your goal in this notebook is to build a useful first dashboard by adding:
* text widgets
* a counter visualization
* a bar chart
* a line chart
* a table visualization
* a pie chart
* one additional chart from the raw `gold_sales` dataset

This notebook is about the UI steps only. It does not introduce new datasets, new SQL, or advanced semantic modeling.

---

## Before You Start

Use the same dashboard you created in the dashboard datasets demo.

Before building visualizations, confirm the following:
1. Your dashboard already contains all six datasets listed above.
2. You are viewing the dashboard draft, not a published read-only version.
3. You are on the **Canvas** tab when adding widgets.
4. Keep the **Data** tab available so you can quickly confirm dataset names if needed.

---

## Step 0: Open the Canvas Tab and Rename the First Page

Before adding any widgets, start by opening the dashboard canvas and giving the first page a clear name.

1. In the dashboard draft, click the page tab named **Untitled Page** to open that page on the canvas.
2. Make sure you are now on the **Canvas** tab for that page.
3. Rename the page so it is easier to understand later.
4. Double click the page name **Untitled Page** and replace it with a clearer name such as **Sales Overview**.
5. Press Enter or click outside the page title field to save the new page name.

---

## Step 1: Add a Title Text Widget

Start the dashboard with a text widget so the page has context before the charts.

1. On the **Canvas**, click **Add a text box**.
2. Click somewhere on the canvas to place the text widget.
3. Double-click the text widget to edit it.
4. Add a dashboard title: `Sales Performance Overview`.
5. Under the title, add a short description: `This dashboard introduces the main sales datasets and highlights revenue, regional performance, trends, and top products.`
7. Adjust the styling of the title and the description.


---

## Step 2: Add a Counter Visualization from `mv_sales_metrics`

Use the metric view dataset for your headline KPI.

1. On the **Canvas**, click **Add a visualization**.
2. Click somewhere on the canvas to place the visualization widget.
2. In the dataset picker, select `mv_sales_metrics`.
3. In the visualization type selector, choose **Counter**.
4. For the primary value, choose the `total_revenue` measure from the metric view.
6. Give the widget a title such as **Total Revenue**.

If you want more KPI cards, repeat the same steps and create separate counters for:
* `Order Count`
* `Avg Order Value`
* `Total Profit`

---

## Step 3: Add a Bar Chart from `star_schema_region_summary`

Use the star schema summary dataset for a region comparison chart.

1. On the **Canvas**, click **Add a visualization**
2. Click somewhere on the canvas to place the visualization widget.
2. Select the dataset `star_schema_region_summary`.
3. Choose **Bar** as the visualization type.
4. Set the x axis to `region_name`.
5. Set the y axis to `total_revenue`.
6. If a color or series field is available and you want a small amount of extra detail, use `territory`. Otherwise leave it blank.
7. Title the widget **Revenue by Region**.

---

## Step 4: Add a Line Chart from `vw_monthly_revenue`

Use the monthly view dataset for a trend chart.

1. On the **Canvas**, click **Add a visualization**
2. Click somewhere on the canvas to place the visualization widget.
2. Select the dataset `vw_monthly_revenue`.
3. Choose **Line** as the visualization type.
4. Set the X-axis to `order_month`.
5. Set the Y-axis to `total_revenue`.
6. Add another line to compare by setting another value in Y-axis to `total_profit`.
7. Title the widget **Monthly Revenue Trend**.

---

## Step 5: Add a Table Visualization from `vw_top_products`

Now add a detailed table so the dashboard includes both summary and detail.

1. On the **Canvas**, click **Add a visualization**
2. Click somewhere on the canvas to place the visualization widget.
2. Select the dataset `vw_top_products`.
3. Choose **Table** as the visualization type.
4. Include columns such as `product_name`, `category`, `total_revenue`, `units_sold`, and `profit_margin_pct`.
5. Reorder the columns so the `category` appears first.
7. Title the widget **Top Products**.

---

## Step 6: Add a Pie Chart from `vw_region_performance`

Use the regional performance view to create a simple part-to-whole visualization.

1. On the **Canvas**, click **Add a visualization**
2. Click somewhere on the canvas to place the visualization widget.
2. Select the dataset `vw_region_performance`.
3. Choose **Pie** as the visualization type.
4. Set the Color to `region_name`.
5. Set the Angle to `total_revenue`.
6. Title the widget **Revenue Share by Region**.

---

## Step 7: Arrange the Dashboard

1. Drag the KPI counter or KPI row to the top.
2. Place the title text widget above or beside the KPI area.
3. Put the bar and pie charts in the next row.
4. Put the line chart below those comparison charts.
5. Place the table across a wider section near the bottom.
6. Resize widgets until the dashboard reads cleanly from top to bottom.

This produces a useful first dashboard without requiring advanced layout features.
