## Why Schedule Refreshes?

Dashboard data doesn't auto-update in real-time. A scheduled refresh:

1. **Pre-caches query results** — viewers get fast load times (data is pre-computed)
2. **Triggers subscriptions** — sends snapshot notifications to subscribers

Without a schedule:
- Data refreshes only when a viewer opens the dashboard (cold query = slow)
- No one gets notified when data changes

With a schedule:
- Results are cached at the scheduled time — viewers get instant loads
- Subscribers receive automatic notifications with the latest snapshot

## Part 1: Set Up a Scheduled Refresh

### Step 1: Open the published dashboard
1. Click **Dashboards** in the left sidebar
2. Find and open: **Sales Dashboard - Scheduling Demo**
3. Make sure you're viewing the **Published** version

### Step 2: Click Schedule
1. Click the **Schedule** button in the upper-right corner (clock icon)
2. A scheduling dialog opens

### Step 3: Configure the schedule
1. **Schedule:** Use the drop-down selectors to specify how often to refresh:
   - Every N hours (e.g., every 6 hours)
   - Every N days (e.g., daily at 8:00 AM)
   - Set when the first run occurs and in which timezone
   - Select **Show cron syntax** for advanced patterns (e.g., weekdays at 8 AM)
2. Click **Advanced settings** to expand additional options:
   - **Name:** Give the schedule a descriptive name (e.g., "Executive Daily")
   - **SQL warehouse:** Optionally choose a different warehouse for this schedule
   - **Use current filter selections:** A checkbox — see explanation below
   - **Custom email subject:** Set a custom subject line for notification emails
   - **Subscribers / Attachments:** Covered in Part 2

### Step 4: Save the schedule
1. Click **Create**
2. The schedule is now active — it will run at the specified time and cadence

> **Cadence note:** For schedules with a defined start time, the refresh cadence resets daily. For example, a schedule set to run every 4 hours starting at 15:10 runs at 15:10, 19:10, 23:10, and so on — then resets at 15:10 the following day.

> **Power BI parallel:** This is like setting up scheduled refresh in the Power BI Service. The key difference: no fixed limit on frequency. You can refresh every hour, every 15 minutes, or once a week.

## Understanding "Use Current Filter Selections"

Inside the schedule's Advanced settings, you'll see a checkbox:

> ☐ **Use current filter selections**

This is easy to overlook, but it's the key to creating **targeted schedules** for different audiences.

### What it does

When you tick this box, the schedule "freezes" whatever filter values are currently applied on the dashboard into this schedule. Every time the schedule runs, it refreshes and snapshots the data **with those filter values applied**.

### The workflow

1. **Set your filters on the dashboard first** (e.g., set Region = "Northeast")
2. Open the Schedule dialog and click **Create new schedule**
3. Tick ☑ **Use current filter selections**
4. The schedule now permanently runs with Region = "Northeast"

Repeat for a different audience:
1. Change the dashboard filter to Region = "Southeast"
2. Create another schedule, tick the box again
3. This schedule runs with Region = "Southeast"

### Why would you use this?

| Scenario | Without the checkbox | With the checkbox |
| --- | --- | --- |
| Executive schedule | Refreshes ALL data (full dashboard) | Refreshes ALL data (same — no filters set) |
| Regional manager schedule | Refreshes ALL data (they see everything) | Refreshes only THEIR region's data |
| Subscription email | Recipient gets the full dashboard PDF | Recipient gets a PDF filtered to their segment |

The real power: **one dashboard, multiple filtered schedules, each sending a tailored snapshot** to different subscriber groups.

### What if you DON'T tick it?

The schedule runs with no filters applied (or with whatever the dashboard's default filter values are). Every subscriber gets the same unfiltered view.

## Bonus: Sharing Dashboards with Pre-Applied Filters via URL

Beyond schedules, you can also share a published dashboard link that already has filters applied:

- When you select filter values on a **published** dashboard, the URL updates to include those selections
- Example: `dashboardsv3/<id>/published?f_<page>~<widget>=Northeast`
- Anyone who opens that URL sees the dashboard with the filter pre-applied
- Users can **bookmark** these filtered views for quick access

This means:
- You can send a teammate a link that opens the dashboard already filtered to their region
- You can create different bookmarks for different filter combinations
- The filters are encoded in the URL — no login-specific state required

## Part 2: Add Subscribers and Configure Attachments

Subscribers receive automatic snapshots when the schedule runs. You configure subscribers and attachments inside the schedule's **Advanced settings**.

### Step 1: Edit your schedule
1. Click the **Schedule** button in the upper-right corner
2. Find your existing schedule and click the three dots to the right and select edit
3. Expand **Advanced settings** if not already visible

### Step 2: Add subscribers
1. Under **Advanced settings -> Subscribers**, add users or notification destinations
2. You can add:
   - **Email addresses** of individual workspace users
   - **Slack channels** (as a notification destination)
   - **Microsoft Teams channels** (as a notification destination)
3. A notification destination counts as one subscriber regardless of channel size

### Step 3: Configure attachments
Expand the **Attachments** section to control what subscribers receive:

1. **Include pages:** Select which dashboard pages to include as PDF attachments (and set the order)
2. **Include data:** Select individual widgets to include as data attachments, and choose format:
   - CSV, TSV, or Excel

### Step 4: Custom email subject (optional)
- Enter a custom subject line so recipients can distinguish between different schedules
- Example: "Weekly Northeast Sales Report" vs "Daily Executive Summary"

### What subscribers receive:

| Channel | What They Get |
| --- | --- |
| **Email** | PDF snapshot of selected pages + optional data attachments (CSV/TSV/Excel) |
| **Slack** | PNG image preview visible in channel + PDF attachment in thread + direct link |
| **Teams** | PNG image preview visible in channel + PDF attachment in thread + direct link |

### Subscriber limits:
- Up to **100 subscribers** per schedule
- A notification destination (Slack channel, Teams channel) counts as **one** subscriber

## Part 3: Create Multiple Schedules (Databricks Advantage)

Unlike Power BI (one schedule per dataset), Databricks supports **multiple schedules per dashboard** — each with different settings.

### Example: Three schedules for different audiences

| Schedule | Frequency | Filter Setup | Subscribers | Purpose |
| --- | --- | --- | --- | --- |
| **Executive Daily** | Daily at 7 AM | No filters applied → checkbox irrelevant | CEO, CFO | Morning briefing (full view) |
| **Regional Weekly** | Monday at 9 AM | Set Region = "Northeast" → tick ☑ Use current filter selections | NE Sales Manager | Weekly regional report |
| **Ops Hourly** | Every 2 hours | No filters applied → checkbox unticked | No subscribers | Cache warming only |

### Step 1: Create the Executive Daily schedule
1. Ensure no filters are selected on the dashboard (clear all)
2. Click **Schedule** → **Create new schedule**
3. Set frequency to daily at 7:00 AM
4. In Advanced settings, give it the name "Executive Daily"
5. Leave "Use current filter selections" **unticked** (no filters to freeze)
6. Add CEO and CFO as subscribers
7. Click **Create**

### Step 2: Create the Regional Weekly schedule
1. On the dashboard, set the Region filter to **"Northeast"**
2. Click **Schedule** → **Create new schedule**
3. Set frequency to weekly, Monday at 9:00 AM
4. In Advanced settings:
   - Name: "NE Regional Weekly"
   - Tick ☑ **Use current filter selections** (this locks in Region = "Northeast")
5. Add the NE Sales Manager as a subscriber
6. Click **Create**
7. The NE manager receives a PDF showing ONLY Northeast data every Monday

### Step 3: Create a cache-warming schedule (no subscribers)
1. Clear filters on the dashboard
2. Click **Schedule** → **Create new schedule**
3. Set frequency to every 2 hours
4. Add NO subscribers — this just pre-caches the data
5. Result: when anyone opens the dashboard, data is already fresh (no cold query)

> **Why this is powerful:** A single dashboard serves multiple audiences with different cadences. The NE manager gets a filtered weekly report. The executive gets the full unfiltered picture daily. The ops team always has fresh cached data. No duplicate dashboards needed — just multiple schedules with different filter snapshots.

## Part 4: Viewer Self-Subscription

Viewers can subscribe themselves to existing schedules without needing the author's help.

### How viewers subscribe:
1. Viewer opens the published dashboard
2. Clicks the **Schedule** button in the upper-right corner
3. Sees all existing schedules listed in the dialog
4. For each schedule, uses the **drop-down menu** to select their preference:

| Option | Behaviour |
| --- | --- |
| **Inactive for me** | Don't participate in this schedule |
| **Refresh data for me** | Pre-cache results for faster load (no notification) |
| **Refresh data for me & email** | Pre-cache + receive email snapshot |

### Why "Refresh data for me" matters:
- Even without receiving a notification, subscribing to a schedule means the cache is populated for that user
- When they open the dashboard, data loads instantly (no cold query)
- This is especially valuable for large datasets where queries take seconds to run

## Manual Refresh

Besides schedules, anyone with **Can Run** permission can manually trigger a refresh:

1. Open the published dashboard
2. Click the **Refresh** button (circular arrow icon, top right)
3. All datasets re-execute against the warehouse
4. Results update in the browser

This is useful for:
- Checking data after a known upstream load completes
- Verifying a fix before the next scheduled refresh
- Ad-hoc exploration when you need the latest numbers

> **Note:** Manual refresh updates the cached results for ALL viewers, not just you.

## Summary

| Feature | What It Does | Power BI Equivalent |
| --- | --- | --- |
| **Scheduled refresh** | Pre-caches data on a cadence | Scheduled dataset refresh |
| **Multiple schedules** | Different frequencies/filters/subscribers per dashboard | Not available (1 schedule per dataset) |
| **Use current filter selections** | Freezes dashboard filters into a schedule for targeted snapshots | Not available |
| **Filter values in URL** | Share a dashboard link with filters pre-applied | URL filter parameters |
| **Email subscription** | PDF snapshot + CSV/TSV/Excel data attachments | Email subscription |
| **Slack/Teams** | PNG preview + PDF attachment to channel | Requires Power Automate |
| **Viewer self-subscribe** | Opt-in to cache refresh and/or notifications | Self-subscribe to emails |
| **Manual refresh** | On-demand re-query | On-demand refresh |
| **Cache warming** | Schedule with no subscribers (just pre-caches) | Not available |


**Best practices:**
- Always set at least one schedule (even without subscribers) for cache warming
- Use multiple schedules to serve different audiences from one dashboard
- Set your filters BEFORE creating a schedule, then tick "Use current filter selections" to lock them in
- Share filtered dashboard URLs with teammates for bookmarkable filtered views
- Use "Refresh data for me" to give power users fast loads without email noise
