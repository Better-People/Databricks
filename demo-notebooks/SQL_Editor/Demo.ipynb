## Step 4: Write and Run Your First Query

### First: Set your default catalog and schema

Before running any query, set your default catalog and schema so you don't have to type the full three-level name (e.g. `catalog.schema.table`) every time.

1. At the **top of the SQL Editor**, you'll see two dropdowns next to the run button — one for **Catalog** and one for **Schema**
2. Click the **Catalog** dropdown and select your workspace catalog
3. Click the **Schema** dropdown and select your schema that starts with `demo_sql_editor_`
4. Now any query you write can reference tables by their short name (e.g. just `sales_summary` instead of `catalog.schema.sales_summary`)

> **Pro tip:** For anything saved or shared, always use the fully qualified three-level namespace. The dropdowns are a convenience for ad-hoc exploration only.

---

### Now: Write your first query

Copy the SQL below into the SQL Editor and run it (click **Run** or press `Ctrl+Enter` / `Cmd+Enter`).

```sql
-- Revenue by region
SELECT 
  region,
  SUM(total_revenue) AS revenue,
  SUM(total_orders) AS orders
FROM sales_summary
GROUP BY region
ORDER BY revenue DESC;
```

### What to notice:

- The results appear in the **Results Panel** at the bottom
- You can resize the results panel by dragging its top edge
- Click on column headers to sort the results
- Use the **Download** button to export results as CSV

> **Pro tip:** If you skip setting the defaults above, you’ll get an error like `TABLE_OR_VIEW_NOT_FOUND`. Just go back and set those catalog/schema dropdowns!

## Step 5: Tab Management

Unlike notebooks where everything is in one vertical scroll, the SQL Editor lets you open **multiple query tabs**.

### Try this:

1. Click the **+** button next to your current tab to and select **Query** create a new query tab
2. Set your default catalog and schema again
3. In the new tab, paste this query:

```sql
-- Top 10 nations by revenue
SELECT 
  nation,
  SUM(total_revenue) AS revenue
FROM sales_summary
GROUP BY nation
ORDER BY revenue DESC
LIMIT 10;
```

3. Run it, then **switch back to your first tab** — notice your previous results are still there
4. You can **rename tabs** by double-clicking the tab title
5. Tabs are **auto-saved** — you won’t lose your work if you close the browser

### Why this matters:

In the SQL Editor, you can have ten queries open side-by-side (well, tab-by-tab) and flip between them instantly.

## Step 6: Multiple Statement Results

Here’s a feature notebooks don’t have: you can run **multiple SELECT statements in one tab** and see all their results.

### Try this:

1. Clear your current tab
2. Paste all three queries below into the **same query editor**:

```sql
-- Query 1: Total revenue
SELECT SUM(total_revenue) AS total_revenue FROM sales_summary;

-- Query 2: Revenue by region
SELECT region, SUM(total_revenue) AS revenue
FROM sales_summary
GROUP BY region
ORDER BY revenue DESC;

-- Query 3: Revenue by month (top 5)
SELECT order_month, SUM(total_revenue) AS revenue
FROM sales_summary
GROUP BY order_month
ORDER BY revenue DESC
LIMIT 5;
```

3. Click **Run** (or `Ctrl+Enter`)
4. Look at the **Results Panel** — you’ll see tabs for each statement’s output
5. Click through the result tabs to compare the different outputs

### Why this matters:

This is perfect for data validation. You can check your totals, your breakdowns, and your edge cases all in one run, then flip through the results to verify everything adds up.

## Step 7: Query Snippets

If you have SQL patterns you use repeatedly, **snippets** let you save them and insert them with a keyword.

### Creating a snippet:

1. In the SQL Editor, decide on the SQL you want to save as a snippet (example below):

```sql
SELECT 
  region,
  nation,
  SUM(total_revenue) AS revenue,
  SUM(total_orders) AS orders,
  ROUND(SUM(total_revenue) / SUM(total_orders), 2) AS avg_order_value
FROM sales_summary
GROUP BY region, nation
ORDER BY revenue DESC
```

2. Click the **⋮ kebab menu** (three dots on the right-hand side of the query editor)
3. Click **View** → **Query snippets**
4. Click **Create query snippet**
5. In the **Replace** field, enter a trigger name like `rns`
6. (Optional) Enter a description
7. In the **Snippet** field, paste the SQL above

8. Click **Create**

### Using a snippet:

10. In a Query tab start typing your trigger name (`rns`)
11. You'll see your snippet appear in the autocomplete — press `Tab` or `Enter` to insert it
12. You can also press `Control+Space` to manually open the autocomplete window and select a snippet

> **Pro tip:** You can add placeholder insertion points in snippets using `${1:placeholder}` syntax. For example, `WHERE amount > ${1:100}` will prompt you to fill in the value when the snippet is inserted.

### Why this matters:

No more copy-pasting from old files or hunting through Slack for that complex join you wrote last month. Snippets keep your reusable SQL at your fingertips.

## Step 8: Auto-Save and Version History

Every query you write in the SQL Editor is **automatically saved** as you type. But what if you make a mistake and want to go back?

### Try this:

1. Make some changes to one of your queries (add a comment, change a column)
2. Click the **⋮** menu (three dots) in the top-right of the query tab
3. Select **Version history**
4. You’ll see a list of previous versions with timestamps
5. Click any version to preview it
6. Click **Restore** to bring back an older version

### Why this matters:

In the SQL Editor, every change is tracked. You can always get back to a working version.

## Step 10: Save Your Query

Once you’ve written a query worth keeping, save it as a named query object.

### Try this:

1. With a query open, click **Save** (or `Ctrl+S`)
2. Give it a descriptive name like `Revenue by Region - Demo 4`
3. Choose a folder location (your home folder is fine)
4. Click **Save**

### What just happened:

- Your query is now a **first-class object** in Databricks
- It appears in the Workspace browser alongside notebooks and files
- You can **share it** with colleagues by adjusting permissions
- It can be **scheduled** to run automatically (we’ll cover this later)
- Other users can **search for it** by name or content

## Summary

You’ve now explored the key features of the SQL Editor:

| Feature | What It Does | 
| --- | --- | --- |
| **Schema Browser** | Browse catalogs, schemas, tables, columns with metadata | 
| **Tab Management** | Work on multiple queries simultaneously | Multiple Power Query Editor windows |
| **Multiple Statement Results** | Run several SELECTs and compare outputs | No direct equivalent |

| **Saved Queries** | Queries as shareable, searchable assets | Queries embedded in .pbix files |

The SQL Editor is your go-to tool for:
- Ad-hoc data exploration
- Writing queries that will power dashboards
- Quick data validation and QA checks
- Building and testing SQL logic before saving as views
