# Demo 3 Setup

This notebook creates the schema and volume used in **Demo 3: Uploading Files via the Databricks UI**.

```python
# Get current user and set up schema name
import re

username = spark.sql("SELECT current_user()").first()[0]
clean_username = re.sub(r'[^a-zA-Z0-9]', '_', username.split('@')[0])
catalog_name = spark.sql("SELECT current_catalog()").first()[0]
schema_name = f"demo_uploads_{clean_username}"
volume_name = "landing_files"

# Create the schema
spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog_name}`.`{schema_name}` COMMENT 'Demo schema for file upload exercises'")
spark.sql(f"USE CATALOG `{catalog_name}`")
spark.sql(f"USE SCHEMA `{schema_name}`")

print(f"\u2713 Schema created: {catalog_name}.{schema_name}")
```

```python
# Create a Volume for the upload-to-volume exercise
spark.sql(f"""
CREATE VOLUME IF NOT EXISTS `{catalog_name}`.`{schema_name}`.`{volume_name}`
COMMENT 'Landing zone for uploaded files - Demo 3'
""")

volume_path = f"/Volumes/{catalog_name}/{schema_name}/{volume_name}"

print(f"\u2713 Volume created: {catalog_name}.{schema_name}.{volume_name}")
print(f"  Path: {volume_path}")
```

```python
# Copy the sample CSV to the volume so students can also access it from there
import shutil, os

source_csv = "/Workspace/Users/" + username + "/sample_sales_data.csv"

# Check if the source file exists in the student's workspace
if os.path.exists(source_csv):
    dest_path = f"{volume_path}/sample_sales_data.csv"
    shutil.copy(source_csv, dest_path)
    print(f"\u2713 Sample CSV copied to volume: {dest_path}")
else:
    print(f"\u26a0 sample_sales_data.csv not found at {source_csv}")
    print(f"  You can still complete the demo by downloading the file from the instructor.")
```

```python
print("")
print("="*60)
print("\u2713 SETUP COMPLETE")
print("="*60)
print(f"")
print(f"Catalog: {catalog_name}")
print(f"Schema:  {schema_name}")
print(f"Volume:  {volume_name}")
print(f"")
print(f"Volume path: {volume_path}")
print(f"")
print(f"You will upload files to:")
print(f"  1. The schema '{schema_name}' (as a new table)")
print(f"  2. The volume '{volume_name}' (as a file)")
```
